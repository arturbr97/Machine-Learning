```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
dados = pd.read_csv('iris.csv')
```

```python
dados
```

```python
dados.loc[dados['species'] == 'setosa','species'] = 0
dados.loc[dados['species'] == 'versicolor','species'] = 1
dados.loc[dados['species'] == 'virginica','species'] = 2
```

```python
X = dados.iloc[:,0:-1]
```

```python
Y = dados.iloc[:,-1]
```

```python
X_treino, X_teste, Y_treino, Y_teste = train_test_split(X,Y,test_size=0.3)
```

```python
MLP = MLPClassifier(hidden_layer_sizes = (100,100), max_iter = 1500 )
MLP.fit(X_treino,Y_treino)
```

```python
valores_preditos_teste = MLP.predict(X_teste)
valores_preditos_treinamento = MLP.predict(X_treino)
```

```python
acuracia_teste = accuracy_score(Y_teste,valores_preditos_teste)
acuracia_treinamento = accuracy_score(Y_treino,valores_preditos_treinamento)

matriz_confusao_teste = confusion_matrix(Y_teste,valores_preditos_teste)

print('Acuracia treino = ', acuracia_treinamento)
print('Acuracia treino = ', acuracia_teste)
print(matriz_confusao_teste)
```
