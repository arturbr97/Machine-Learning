```python
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
Boston = load_boston()
```

```python
Atributos = Boston.data
```

```python
Precos = Boston.target
```

```python
plt.xlabel('Atributos')
plt.ylabel('Preco')
plt.title('Atributos X Preco')
plt.plot(Atributos,Precos,'o')
plt.axis([0,100,0,100])
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo_regressao_linear = LinearRegression()
```

```python
Atributos.reshape(-1,13)
modelo_regressao_linear.fit(Atributos,Precos)
```

```python
modelo_regressao_linear.predict(Atributos[0].reshape(-1,13))
```

```python
plt.figure()
plt.xlabel('Diametro')
plt.ylabel('Preco')
plt.title('Atributos X Preco')
plt.plot(Atributos,Precos,'o')
plt.plot(Atributos,modelo_regressao_linear.predict(Atributos),linewidth=3)
plt.axis([0,60,0,60])
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.metrics import mean_squared_error
```

```python
degree = 3

modelo_polinomial = make_pipeline(PolynomialFeatures(degree), Ridge())
modelo_polinomial.fit(Atributos, Precos)
```

```python
print("Mean squared error = ", mean_squared_error(Precos,modelo_polinomial.predict(Atributos)))
```

```python
print(mean_squared_error(Precos,modelo_regressao_linear.predict(Atributos)))
print(mean_squared_error(Precos,modelo_polinomial.predict(Atributos)))
```
